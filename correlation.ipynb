```python
import pandas_datareader as pdr

aapl = pdr.get_data_yahoo('AAPL')
msft = pdr.get_data_yahoo('MSFT')
```

```python
import datetime

def daily_returns(df):
    dr = df['Close'][1:] / df['Close'][:-1].values - 1
    dr[datetime.datetime(2009, 12, 31)] = 0
    
    return dr
```

```python
aapl['Daily Returns'] = daily_returns(aapl)
msft['Daily Returns'] = daily_returns(msft)
```

```python
import datetime

df = aapl.join(msft, lsuffix=' AAPL', rsuffix=' MSFT', sort=True)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr()
sns.heatmap(corr)
plt.show()

corr
```
