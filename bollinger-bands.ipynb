```python
import datetime
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import pandas as pd
```

```python
df0 = pdr.get_data_yahoo('GOOG')
```

```python
df0.head()
```

```python
df0.tail()
```

```python
open_ = df0['Open']

open_mean = open_.rolling(20).mean()
open_std = open_.rolling(20).std()
open_high = open_mean + 2 * open_std
open_low = open_mean - 2 * open_std

df1 = pd.DataFrame(data={
    'Value': open_,
    'Mean': open_mean,
    'High': open_high,
    'Low': open_low
})['2010-01-29':]
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1['Value'].plot()
df1['Mean'].plot()
df1['Low'].plot()
df1['High'].plot()
plt.show()
```

```python
df2 = df1['2015-01-01':'2016-01-01']

df2['Value'].plot()
df2['Mean'].plot()
df2['Low'].plot()
df2['High'].plot()
plt.show()
```

```python
df3 = df1['2015-06-01':'2015-10-01']

df3['Value'].plot()
df3['Mean'].plot()
df3['Low'].plot()
df3['High'].plot()
plt.show()
```
